### Question 1

```python
def curve(A,B,C):
    t = np.linspace(0,2*np.pi,1000)
    x = A*np.sin(t)**3
    y = B*np.cos(t) - C*np.cos(2*t)
    plt.plot(x,y), plt.axis('equal')
```

```python
curve(13,6,4)
```

### Question 2

```python
def power_series(a,x):
    N = len(a)
    xs = np.array([x**k for k in range(0,N)])
    return np.sum(a*xs)
```

```python
power_series(np.ones(100),0.5)
```

```python
from scipy.special import factorial
```

```python
power_series(1 / factorial(np.arange(0,100)),1)
```

### Question 3

```python
def triangle_wave(N,interval):
    x = np.linspace(interval[0],interval[1],(interval[1] - interval[0])*100)
    y = (8/np.pi**2) * sum([(-1)**k * np.sin(2*np.pi*(2*k+1)*x) / (2*k+1)**2 for k in range(0,N+1) ])
    plt.plot(x,y)
    plt.ylim([-2,2])
    plt.grid('on')
```

```python
triangle_wave(3,[0,3])
```

### Question 4

```python
def sawtooth_waves(n,m,T):
    t = np.linspace(0,T,100*T)
    plt.figure(figsize=(2*m,2*n))
    plt.subplot(n,m,1)
    for i in range(1,n*m+1):
        plt.subplot(n,m,i)
        y = (2/np.pi) * sum([(-1)**(k+1) * np.sin(2*np.pi*k*t)/k for k in range(1,i+1)])
        plt.plot(t,y)
        plt.title('N = {}'.format(i))
        plt.ylim([-1.5,1.5])
        plt.grid('on')
        plt.tight_layout()
    plt.show()
```

```python
sawtooth_waves(2,4,2)
```

### Question 5

```python
def is_prime(N):
    if N < 2:
        return False
    else:
        for d in range(2,int(N**0.5) + 1):
            if N % d == 0:
                 return False
        return True
```

```python
def euler_product(s,N):
    primes = [n for n in range(2,N + 1) if is_prime(n)]
    return np.array([(1 - p**(-s))**(-1) for p in primes]).prod()
```

```python
euler_product(4,10000)
```

### Question 6

```python
def slope_field(f,xlims,ylims,grid_step):
    xs = np.arange(xlims[0],xlims[1],grid_step)
    ys = np.arange(ylims[0],ylims[1],grid_step)
    L = grid_step * 0.4
    for x in xs:
        for y in ys:
            m = f(x,y)
            theta = np.arctan(m)
            dx = L * np.cos(theta)
            dy = L * np.sin(theta)
            plt.plot([x-dx,x+dx],[y-dy,y+dy],'b')
            plt.hold('on')
    plt.xlim(xlims), plt.ylim(ylims)
    plt.show()
```

```python
def f(t,y):
    return y**2 - t**2
```

```python
slope_field(f,[-3,3],[-2,2],0.2)
```
