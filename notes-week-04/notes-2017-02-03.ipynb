## 1. More plotting examples

Let's plot some more examples using NumPy arrays and matplotlib's pyplot commands. See the [documentation](http://matplotlib.org/) for more details.

### Example: Trefoil knot in 2D

The [trefoil knot](https://en.wikipedia.org/wiki/Trefoil_knot) in 2D is given by the parametric equations:

$$
x(t) = \sin(t) + 2\sin(2t) \ \ , \ \ y(t) = \cos(t) - 2 \cos(2t) \ \ , \ t \in [0,2 \pi]
$$

```python
t = np.linspace(0,2*np.pi,1000)
x = np.sin(t) + 2*np.sin(2*t)
y = np.cos(t) - 2*np.cos(2*t)
```

```python
plt.plot(x,y)
plt.axis('equal')
plt.xlim([-4,4])
plt.ylim([-4,3])
plt.title('Trefoil knot in 2D');
```

### Example: Plot several curves together

The `plot` command can take several paris of arrays `x` and `y` to plot several curves in teh same figure. For exampl, plot the curve $y = e^{-x^2} \cos(k x)$ for $k = 1,2,5,10$ for $x \in [-5,5]$.

```python
x = np.linspace(-5,5,1000)
y1 = np.exp(-x**2) * np.cos(x)
y2 = np.exp(-x**2) * np.cos(2*x)
y3 = np.exp(-x**2) * np.cos(5*x)
y4 = np.exp(-x**2) * np.cos(10*x)
plt.plot(x,y1,x,y2,x,y3,x,y4);
```

Alternatively, we can call the `plot` command several times in a for loop and each plot will be added to the figure until the cell code is complete and the figure is rendered in the notebook.

```python
x = np.linspace(-2,2,1000)
ks = np.random.rand(100)*10
for k in ks:
    y = np.exp(-x**2) * np.cos(k*x)
    plt.plot(x,y);
```

## 2. Array methods and attributes

[NumPy arrays](https://docs.scipy.org/doc/numpy/reference/arrays.html) have many methods (ie. functions) to transform them and to do calculations. We access methods using the dot notation following the array variable name. See the [documentation](https://docs.scipy.org/doc/numpy/reference/arrays.ndarray.html#array-methods) to see a list of array methods. For example, there is the `sum` method which simply sums the entries in the array.

```python
arr = np.array([1,2,3,4,5])
```

```python
arr.sum()
```

There is an equivalent functional way of doing the same thing:

```python
np.sum(arr)
```

There is also the prod method which takes the product of the entries:

```python
arr.prod()
```

We can use this to make a factorial function:

```python
def factorial(n):
    return np.array(range(1,n+1)).prod()
```

```python
factorial(3)
```

```python
factorial(5)
```

The method `cumsum` computes the cumulative sum of the entries:

```python
arr.cumsum()
```

For example, we can plot a random walk:

```python
x = np.random.randn(1000,10).cumsum(axis=0) # Take the cumulative sum down the columns
plt.plot(x); # Plot each of the 10 columns of x
```

There is also the `transpose` method:

```python
A = np.array([[1,2],[3,4]])
A
```

```python
A.transpose()
```

Also, there's the `mean`:

```python
A.mean()
```

```python
A.mean(axis=0) # Axis 0 is across the rows
```

```python
A.mean(axis=1) # Axis 1 is down the columns
```

```python
A
```

NumPy arrays also have [attributes](https://docs.scipy.org/doc/numpy/reference/arrays.ndarray.html#array-attributes) which are values intrinsic to the array such as its shape. We accessed attributes with the dot notation but with the parentheses `()` since these are values and not functions like methods.

For example, we can access the shape of an array using `.shape`:

```python
M = np.random.randint(0,2,[4,5])
print(M)
M.shape
```

## 3. Array slicing and indexing

Accessing the entries in an array is called [indexing](https://docs.scipy.org/doc/numpy/reference/arrays.indexing.html) and accessing rows and columns (or subarrays) is called slicing. For example, we access the entries in an array using the square brackets:

```python
A = np.random.randint(0,10,[3,5])
A
```

```python
A[2,4] # Row index 2 and column index 4
```

```python
A[0,0]
```

An example of slicing is to access a whole row (or column) using the colon `:`

```python
A[0,:]
```

```python
A[:,3]
```
