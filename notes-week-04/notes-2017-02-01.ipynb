### Array operations

We can do arithmetic operations on arrays like addition, multiplication and division which are performed elementwise. For example, array addition/subtraction is the same as vector addition/subtraction as in linear algebra:

```python
v = np.array([1,2,3])
w = np.array([1,0,-1])
```

```python
v + w
```

```python
v - w
```

In the same way, array multiplication/division is performed element by element:

```python
v * w
```

```python
w / v
```

The exponent operator `**` also acts element by element in the array:

```python
u = np.arange(0,5,0.5)
u
```

```python
u ** 2
```

## 2. Mathematical Functions in NumPy

NumPy has **many** mathematical functions built-in. These functions are special kinds of functions called "Universal Functions" or `ufuncs`. (In other languages, this kind of function is also called a *vectorized* function.)

Universal functions are built to compute values across arrays very quickly. This means that when you pass in an array of values into a `ufunc`, the function will compute the value of the function for every entry in the array.

## 3. Plotting with Maplotlib

The basic plotting function in Matplotlib is `matplotlib.pyplot.plot`, or `plt.plot` when we import `matplotlib.pyplot` as `plt`. The basic procedure is:

1. Make an array of x values
2. Use the array of x values to create an array of y values
3. Enter `plt.plot(x,y)`
4. Add style to the plot using various `pyplot` commands such as `plt.title`, `plt.xlim`, etc...

Check out the `pyplot` [documentation](http://matplotlib.org/api/pyplot_summary.html) for more options.

### Example: Step-by-step instructions to plot a polynomial function

Let's plot the graph of the function $f(x) = x^4-4x^2-x+1$ on the interval $[-2,2]$. First, we generate the NumPy array of $x$ values:

```python
x = np.linspace(-2,2,100)
```

And then we use the array `x` to create the array of `y` values. Remember, all arithmetic operations such as `+`, `*`, `**` and `/` are executed element by element. 

```python
y = x**4 - 4*x**2 - x + 1
```

Let's plot the result:

```python
plt.plot(x,y);
```

Finally, let's plot the polynomial again but now with some style.

```python
plt.plot(x,y,'ro',markersize=2)
plt.title('Graph of y = x^4 - 4x^2 - x + 1')
plt.grid('on')
plt.xlabel('x')
plt.ylabel('y');
```

### Example: Plot a parabola

Let's plot the graph of $y = 1 - x^2$ for $x \in [-2,2]$:

```python
x = np.linspace(-2,2,100)
y = 1 - x**2
plt.plot(x,y,'r*'); # Each point is a red star
```

### Example: Plot a sine wave

Let's plot the graph of $y = \sin(x)$ for $x \in [-50,50]$:

```python
x = np.linspace(-50,50,500)
y = np.sin(x)
plt.plot(x,y,'b.') # Plot each point as a blue dot
plt.xlim([-50,50]), plt.ylim([-2,2]); # Set the x and y limits of the display
```

### Example: Plot a decaying cosine wave

Let's plot $y = e^{-x^2/10} \cos(5x)$ for $x \in [-10,10]$:

```python
x = np.arange(-10,10,0.01)
y = np.exp(-x**2 / 10) * np.cos(5*x)
plt.plot(x,y), plt.ylim([-1.5,1.5]), plt.grid('on');
```

### Example: Plot a parametric curve

Let's plot a circle: $x = \cos(t)$ and $y = \sin(t)$ for $t \in [0,2\pi]$

```python
t = np.linspace(0,2*np.pi,100)
x = np.cos(t)
y = np.sin(t)
plt.figure(figsize=[5,5])
plt.plot(x,y)
plt.axis('equal'), plt.xlim([-2,2]), plt.ylim([-2,2]);
```

### Example: Plot a spiral

Let's plot $x = t \cos(10t)$ and $y = t \sin(10t)$ for $t \in [0,2\pi]$:

```python
t = np.linspace(0,2*np.pi,1000)
x = t * np.cos(10*t)
y = t * np.sin(10*t)
plt.plot(x,y), plt.axis('equal'), plt.xlim([-2,2]), plt.ylim([-2,2]);
```

### Example: Plot the butterfly curve

Let's plot the butterfly curve!

![butterfly](https://upload.wikimedia.org/wikipedia/commons/2/2a/Butterfly_trans01.svg)

(Source: Wikipedia - [Butterfly curve!](https://en.wikipedia.org/wiki/Butterfly_curve_%28transcendental%29))

```python
t = np.linspace(0,20*np.pi,3000)
u = np.exp(np.cos(t)) - 2*np.cos(4*t) - np.sin(t/12)**5
x = np.sin(t) * u
y = np.cos(t) * u
plt.plot(x,y,'g')
plt.axis('off')
plt.axis('equal');
```
