## 1. Exploring `pyplot` commands

There are many [`pyplot`](http://matplotlib.org/api/pyplot_summary.html) commands available for us to customize our figures. For example:

* `xlim` and `ylim` set the display limits of the figure
* `grid` adds gird lines
* `title` adds a title to the figure
* `xlabel` and `ylabel` add labels to the horizontal and vertical axes
* `savefig` to save the figure (as .png, .pdf, etc) to the current working directory
* `show` renders the plot (this command is implicitly called by the Jupyter notebook when a cell is finished executing)
* `legend` display a legend for several lines in the same figure

The best way to explore these commands is by examples!

### Example 1

Let's plot $y = \sin(x) + \sin(4x)$ for $x \in [0,2\pi]$ and add style to our plot.

```python
# Create x and y arrays for our plot
x = np.linspace(0,2*np.pi,1000)
y = np.sin(x) + np.sin(4*x)

# Plot the curve add style to line
plt.plot(x,y,'r--',linewidth=2)

# Set the x and y limits of the display, and add dash-dot grid lines
plt.xlim([0,2*np.pi]), plt.ylim([-3,3]), plt.grid('on',linestyle='-.')

# Add title and axes labels
plt.title('Sum of sine waves'), plt.xlabel('Time (s)'), plt.ylabel('Amplitude (mm)')

# Save
plt.savefig('sine.png',format='png')
```

And then we can insert the image in the notebook using markdown:

> `![New Figure](sine.png)`

displays the figure: ![New Figure](sine.png)

### Example 2

There's also a special command called `show`. We use this when we want to render the current figure. Jupyter notebook calls this command when it reaches the end of a cell with a figure. We can use the `show` command to render several figures from within the same cell.

```python
x = np.linspace(0,2,1000)
y1 = np.sin(2*np.pi*x) + np.sin(4*np.pi*x)
plt.plot(x,y1), plt.title('Figure 1')
plt.show()

y2 = np.sin(2*np.pi*x) + np.sin(4*np.pi*x) + np.sin(6*np.pi*x)
plt.plot(x,y2,'r'), plt.title('Figure 1')
plt.show()
```

### Example 3

We can also generate a legend for a figure with several lines. For example, let's plot $y = \sin(x)$ and $y = \cos(x)$ in the same figure with a legend.

```python
t = np.linspace(0,4*np.pi,100)
y1 = np.sin(t)
y2 = np.cos(t)
plt.plot(t,y1,t,y2), plt.xlim([0,4*np.pi]), plt.ylim([-2,2])
plt.legend(['y = sin(t)','y = cos(t)']);
```

## 2. Subplot

The command `subplot` takes at least 3 inputs $n$, $m$ and $i$ and creates a figure with a $n$ by $m$ grid of subplots and then sets the $i$th subplot (counting across the rows) as the current plot (ie. current axes object).

For example, let's plot the curve

$$
y = \frac{x^{k} \arctan(x)}{1 + x^2} \ , \ k = 0, 1, 2, 3 \ , \ x \in [-5,5]
$$

```python
for k in range(1,5):
    plt.subplot(2,2,k)
    x = np.linspace(-5,5,1000)
    y = x**(k-1) * np.arctan(x) / (1 + x**2)
    plt.plot(x,y), plt.grid('on'), plt.xlim([-5,5])
    plt.title('k = ' + str(k-1)), plt.tight_layout()
```

## 3. Different kinds of plots

So far we've just been creating line plots. Let's create some other kinds.

### Scatter plots

```python
N = 10000
x = np.random.randn(N)
y = np.random.rand(N)
plt.scatter(x,y,lw=0,alpha=0.2,c='green');
```

### Bar plots
